```python
df.loc[df["sex"]=="male","sex"]=1  #1->male
df.loc[df["sex"]=="female","sex"]=0  #0->female
```

```python
df.loc[df["smoker"]=="yes","smoker"]=1  #1->is_smoker
df.loc[df["smoker"]=="no","smoker"]=0  #0->not_smoker
```

```python
status=pd.get_dummies(df['region'],drop_first=True)
df=pd.concat([df,status],axis=1)
df.head()
```

```python
from sklearn.model_selection import train_test_split
X=df[["age","sex","bmi","children","smoker",'northwest','southeast','southwest']]
y=df[["charges"]]
```

```python
X.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=5)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
lm.fit(X_train,y_train)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
y_test_pred=lm.predict(X_test)
```

```python
plt.scatter(y_test,y_test_pred)
```

```python
y_train_pred=lm.predict(X_train)
```

```python
plt.scatter(y_train,y_train_pred)
```

```python
res=y_train-y_train_pred
sns.distplot(res)
```

```python
score = lm.score(X_test,y_test)
score*100
```

```python
from sklearn.metrics import r2_score
r2_score(y_train,y_train_pred)
```

```python
r2_score(y_test,y_test_pred)
```
