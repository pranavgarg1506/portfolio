```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn import preprocessing, tree, ensemble, model_selection, metrics, linear_model
```

```python
def logistic(x,y):
    x_train, x_test, y_train, y_test = model_selection.train_test_split(x,y,test_size = 0.2,random_state=7)
    clf = linear_model.LogisticRegression()
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    #scores = model_selection.cross_val_score(clf, x,y,cv=5,scoring='accuracy')
    #print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
    
    #scores1 = model_selection.cross_validate(clf,x,y,cv=5,scoring=["accuracy"],return_estimator=True)
    #print(scores1)
    print("Accuracy in Logistic Regression",metrics.accuracy_score(y_test, y_pred))
```

```python
def randomForest(x,y):
    x_train, x_test, y_train, y_test = model_selection.train_test_split(x,y,test_size = 0.2,random_state=7)
    clf = ensemble.RandomForestClassifier()
    clf.fit(x_train,y_train)
    y_pred = clf.predict(x_test)
    
    print("Accuracy in Random Forest",metrics.accuracy_score(y_test, y_pred))
```
